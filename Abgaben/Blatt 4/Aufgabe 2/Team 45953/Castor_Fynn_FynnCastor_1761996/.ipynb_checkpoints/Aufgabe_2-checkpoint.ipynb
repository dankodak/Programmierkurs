```python

def heaviside(x):
    if x<0:
        return 0
    else:
        return 1
    



def smoothed_heavyside(x,epsilon=1*10**-3):
    from math import sin, pi
    if epsilon<=0:
        print("Error: please type in a positive epsilon")
    else:
        if x<-epsilon:
            return 0
        if x>=-epsilon and x<=-epsilon:
            h=0.5*(x/(epsilon)+1+sin(pi+x/epsilon)/pi)
            return h
        if x>epsilon:
            return 1
    



def get_piecewise_polynomial(p,a):
    r=0
    if isinstance(p,list) and isinstance(a,list) is False:
        print("Error: p or a is not a list")
        r=1
    
    elif len(a)+1!=len(p):
        print("Error: p is not 1 element larger than a")
        r=1
    
    j=p[0]-1
    
    for i in p:
        if j>=i:
            print("Error: pn is smaller than pn-1")
            break
        j=i
    if r==0:
        def g(x):
        
            if x>=p[0] and x<p[len(p)-1]:
                k=0
            
                while x>=p[k]:
                    k+=1
               
                h=k-1
                return a[h]
            else :
                return 0
        return g
```
