Aufgabe 2: Matrixmultiplikation mit numpy Matrizen und Listen

```python
#(a) Multiplikation 'von Hand':

def matrix_multiplication(A,B):
    C = []
    
    for i in range(len(A)):
        row = []
        for j in range(len(B)):
            sum = 0
            for k in range(len(B[0])):
                sum = sum+(A[i][k]*B[k][j])
            row.append(sum)
        C.append(row)
        row = []
    return C

A = [[1,2,3,4],
     [3,4,5,6],
     [5,6,7,8],
     [7,8,9,10]]

B = [[3,4,5,6],
     [5,6,7,8],
     [7,8,9,10],
     [9,10,11,12]]

print(matrix_multiplication(A,B))

#(b) Matrixmultiplikation mit numpy und Zeitvergleich:
import numpy as np
import time

A = np.ones((100, 100))
N = 10

start1 = time.clock()
for i in range(N):
    matrix_multiplication(A,A)
end1 = time.clock()
time1 = end1 - start1
time1_sin = time1/N
print(matrix_multiplication(A,A))

print(("von Hand: " +
       "Gesamtzeit: {} Sek.\n" +
     "ein Schleifendurchlauf: {} Sek.").format(time1, time1_sin))

start2 = time.clock()
for i in range(N):
    np.dot(A,A)
end2 = time.clock()
time2 = end2 - start2
time2_sin = time2/N
print(np.dot(A,A))

print(("numpy: " +
       "Gesamtzeit: {} Sek.\n" +
     "ein Schleifendurchlauf: {} Sek.").format(time2, time2_sin))
```
